```python
# class 갯수 측정. 시각화
"""
'Adenopathy', 'Atelectasis',
       'Azygos Lobe', 'Calcification of the Aorta', 'Cardiomegaly',
       'Clavicle Fracture', 'Consolidation', 'Edema', 'Emphysema',
       'Enlarged Cardiomediastinum', 'Fibrosis', 'Fissure', 'Fracture',
       'Granuloma', 'Hernia', 'Hydropneumothorax', 'Infarction',
       'Infiltration', 'Kyphosis', 'Lobar Atelectasis', 'Lung Lesion',
       'Lung Opacity', 'Mass', 'Nodule', 'Normal', 'Pleural Effusion',
       'Pleural Other', 'Pleural Thickening', 'Pneumomediastinum', 'Pneumonia',
       'Pneumoperitoneum', 'Pneumothorax', 'Pulmonary Embolism',
       'Pulmonary Hypertension', 'Rib Fracture', 'Round(ed) Atelectasis',
       'Subcutaneous Emphysema', 'Support Devices', 'Tortuous Aorta',
       'Tuberculosis']
"""

import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(20,10))
class_count = train_df.iloc[:,7:].sum()
class_count_sorted = class_count.sort_values(ascending=False)

# Set the figure size
plt.figure(figsize=(20, 10))

# Create the bar plot
sns.barplot(x=class_count_sorted.index, y=class_count_sorted.values, alpha=0.8)

# Rotate the class names on the x-axis
plt.xticks(rotation=90)

# Set labels and title for better readability
plt.xlabel('Class Names')
plt.ylabel('Counts')
plt.title('Class Distribution')

# Display the plot
plt.show()
```
